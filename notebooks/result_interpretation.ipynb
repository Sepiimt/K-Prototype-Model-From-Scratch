```python
import numpy as np
import pandas as pd
import csv
import sys

#> --- API Imports ---
try:
    sys.path.insert(0, "..\\src\\k_prototype")
    from k_prototype import in_depth_decoder
except:
    sys.path.insert(0, "..\\src\\")
    from k_prototype import in_depth_decoder
```

```python
numerical_prototypes = np.expm1(np.load(r'..results\encoded_and_seperated_model_prototypes\numerical_prototypes.npy'))
categorical_prototypes = np.load(r'..results\encoded_and_seperated_model_prototypes\categorical_prototypes.npy')

location_decoder_map = np.load(r'..\\data\\decoder_maps\\location_decoder_map.npy', allow_pickle=True)
rest_type_decoder_map = np.load('..\\data\\decoder_maps\\rest_type_decoder_map.npy', allow_pickle=True)
cuisines_decoder_map = np.load('..\\data\\decoder_maps\\cuisines_decoder_map.npy', allow_pickle=True)
listed_in_type_decoder_map = np.load('..\\data\\decoder_maps\\listed_in_type_decoder_map.npy', allow_pickle=True)
listed_in_city_decoder_map = np.load('..\\data\\decoder_maps\\listed_in_city_decoder_map.npy', allow_pickle=True)
```

```python
# in_depth_decoder(encoded_codes, decoder_mapping)
decode_maps = [location_decoder_map, rest_type_decoder_map, cuisines_decoder_map, listed_in_type_decoder_map, listed_in_city_decoder_map]

decoded_categorical_prototypes = []
decoded_categorical_prototypes.append(np.where(categorical_prototypes[:,0]==1, "Yes", "No"))
decoded_categorical_prototypes.append(np.where(categorical_prototypes[:,1]==1, "Yes", "No"))

for column, decode_maps in zip(range(2,7), decode_maps):
    decoded_categorical_prototypes.append(in_depth_decoder(categorical_prototypes[:,column], decode_maps))
    
decoded_categorical_prototypes = np.array(decoded_categorical_prototypes).T
```

```python
decoded_categorical_prototypes_data_frame = pd.DataFrame(decoded_categorical_prototypes)
numerical_prototypes_data_frame = pd.DataFrame(numerical_prototypes)
prototypes_data_frame = pd.concat([numerical_prototypes_data_frame, decoded_categorical_prototypes_data_frame], axis=1)
prototypes_data_frame.columns = ["rate", "votes", "approx_cost(for two people)", "online_order", "book_table", "location", "rest_type", "cuisines", "listed_in(type)", "listed_in(city)"]
```

```python
prototypes_data_frame.to_csv(r'..results\decoded_model_prototypes\prototypes_data_frame.csv', index=False)
```
